```python
import numpy as np
import pandas as pd
import scipy
```

```python
df = pd.read_csv('heart.csv')
print(df.describe())
print(df.shape)
```

```python
#get z Zscore
```

```python
numricDf = df.select_dtypes(include=np.number)
from scipy import stats
Zscore = np.abs(stats.zscore(numricDf))
Zscore
                
```

```python
# now remove outlier using zscore above 3
```

```python
count = 3
cleanDf = df[(Zscore < count).all(axis = 1)] #= df[(z_scores < 3).all(axis=1)]
```

```python
cleanDf.shape
```

```python
cleanDf.head()
```

```python
from sklearn.preprocessing import LabelEncoder
label = LabelEncoder()
for col in ['Sex','ChestPainType','RestingECG','ExerciseAngina','ST_Slope']:
    cleanDf[col]= label.fit_transform(cleanDf[col])
```

```python
cleanDf.head()
```

```python
X =  cleanDf.drop("HeartDisease",axis=1)
X.shape
```

```python
y = cleanDf.HeartDisease
y.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_Scaled = scaler.fit_transform(X)
X_Scaled
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test ,y_train , y_test = train_test_split(X_Scaled,y,random_state=30)
```

```python
from sklearn.svm import  SVC
model = SVC()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
from sklearn.ensemble import BaggingClassifier
bagmodel = BaggingClassifier(estimator=SVC(),n_estimators=100,max_samples=.8,oob_score=True,random_state=30)
bagmodel.fit(X_train,y_train)
bagmodel.oob_score_
```

```python
from sklearn.tree import DecisionTreeClassifier
model_tree = DecisionTreeClassifier()
model_tree.fit(X_train,y_train)
```

```python
model_tree.score(X_test,y_test)
```

```python
from sklearn.ensemble import BaggingClassifier
bagmodel_tree = BaggingClassifier(estimator=DecisionTreeClassifier(),n_estimators=100,max_samples=.8,oob_score=True,random_state=30)
bagmodel_tree.fit(X_train,y_train)
bagmodel_tree.oob_score_
```
